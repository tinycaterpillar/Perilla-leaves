# 음성의 MFCC를 관찰하는 프로그램
좋은 feature 인듯

```python
import pandas as pd
from IPython.display import Audio, display
from pydub import AudioSegment
import numpy as np
import matplotlib.pyplot as plt
import librosa
```

```python
seed = 42
# seed = 123
# seed = 9903
train_file = "train.csv"
test_file = "test.csv"
```

```python
train_df = pd.read_csv(train_file)
test_df = pd.read_csv(test_file)
train_df.head(3)
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# 오디오 파일 로드
filename = './train/JFAWUOGJ.ogg'
y, sr = librosa.load(filename)

# MFCC 계산
n_mfcc = 20  # MFCC 계수의 수
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)

# 델타 MFCC 계산
delta_mfcc = librosa.feature.delta(mfcc)

# 델타-델타 MFCC 계산
delta2_mfcc = librosa.feature.delta(mfcc, order=2)

# MFCC 시각화
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
librosa.display.specshow(mfcc, sr=sr, x_axis='time')
plt.colorbar()
plt.title('MFCC')

plt.subplot(3, 1, 2)
librosa.display.specshow(delta_mfcc, sr=sr, x_axis='time')
plt.colorbar()
plt.title('Delta MFCC')

plt.subplot(3, 1, 3)
librosa.display.specshow(delta2_mfcc, sr=sr, x_axis='time')
plt.colorbar()
plt.title('Delta-Delta MFCC')

plt.tight_layout()
plt.show()
```

```python
train_sample_real_df = train_df[train_df['label'] == "real"].sample(n=5, random_state=seed)
train_sample_fake_df = train_df[train_df['label'] == "fake"].sample(n=5, random_state=seed)
test_sample_df = test_df.sample(n=5, random_state=seed)
```

## train real  
방음 환경에서의 진짜 사람 목소리와 가짜 생성 목소리   
각 샘플 당 한명의 목소리

```python
for index, row in train_sample_real_df.iterrows():
    # 오디오 파일 로드
    filename = row['path']
    y, sr = librosa.load(filename)

    # MFCC 계산
    n_mfcc = 20  # MFCC 계수의 수
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)

    # 델타 MFCC 계산
    delta_mfcc = librosa.feature.delta(mfcc)

    # 델타-델타 MFCC 계산
    delta2_mfcc = librosa.feature.delta(mfcc, order=2)

    # MFCC 시각화
    plt.figure(figsize=(10, 6))

    plt.subplot(3, 1, 1)
    librosa.display.specshow(mfcc, sr=sr, x_axis='time')
    plt.colorbar()
    plt.title('MFCC')

    plt.subplot(3, 1, 2)
    librosa.display.specshow(delta_mfcc, sr=sr, x_axis='time')
    plt.colorbar()
    plt.title('Delta MFCC')

    plt.subplot(3, 1, 3)
    librosa.display.specshow(delta2_mfcc, sr=sr, x_axis='time')
    plt.colorbar()
    plt.title('Delta-Delta MFCC')

    plt.tight_layout()
    plt.show()
```

## train fake
