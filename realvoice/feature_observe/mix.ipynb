# 데이터를 mixing 하는 프로그램

```python
import pandas as pd
from IPython.display import Audio, display
from pydub import AudioSegment, effects  
```

```python
seed = 42
# seed = 123
# seed = 9903
train_file = "train.csv"
test_file = "test.csv"
```

```python
train_df = pd.read_csv(train_file)
test_df = pd.read_csv(test_file)
train_df.head(3)
```

```python
audio_file1 = "./train/RUNQPNJF.ogg"
audio_file2 = "./train/JFAWUOGJ.ogg"

print(f"Playing audio file: {audio_file1}")
display(Audio(audio_file1))
print(f"Playing audio file: {audio_file2}")
display(Audio(audio_file2))
```

```python
# 두 OGG 파일 불러오기
audio1 = AudioSegment.from_file(audio_file1, format="ogg")
audio2 = AudioSegment.from_file(audio_file2, format="ogg")
audio1 = effects.normalize(audio1)  
audio2 = effects.normalize(audio2)  

# 두 음원 파일을 동시에 재생되도록 합치기 (믹싱)
combined = audio1.overlay(audio2)

display(combined)
```

```python
# 저장
combined.export("sample.ogg", format="ogg")
```
